```python
def log_decorator(func):
  def wrap():
    print(f"Calling func {func}")
    func()
    print(f"Func {func} finished if's work")
  return wrap  
```

```python
def hello():
  print("Hello, world")
```

```python
wrapped_by_logger = log_decorator(hello)
wrapped_by_logger()
```

```python
@log_decorator
def hello():
  print("hello, world")
```

```python
hello()
```

```python
from timeit import default_timer as timer
import math
import time
```

```python
def measure_time(func):
  def inner(*args, **kwargs):
    start = timer()

    func(*args, **kwargs)

    end = timer()

    print(f"Function {func.__name__} took {end-start} for execution")
  return inner
```

```python
@measure_time
def factorial(num):
  time.sleep(3)
  print(math.factorial(num))
```

```python
factorial(100)
```
