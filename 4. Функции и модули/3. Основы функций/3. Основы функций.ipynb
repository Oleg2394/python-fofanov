```python
def is_pallindrom(text): # С начала и с конца все одинаковое
  return text == text[::-1]

print(is_pallindrom("aabaa"))  
print(is_pallindrom("aabba"))  
```

```python
def calc_taxes(p1, p2, p3):
  return sum((p1,p2,p3)) * 0.06
```

```python
calc_taxes(10,20,30)
```

```python
calc_taxes(10,20,30,40)
```

```python
def calc_taxes(*args): # Принимать любое количество аргументов
  for x in args:
    print(f" Got payment = {x}")
  return sum(args) * 0.06  
```

```python
calc_taxes(10,20,30,40)
```
