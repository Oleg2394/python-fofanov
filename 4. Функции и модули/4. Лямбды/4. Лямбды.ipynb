```python
def square(*args):
  return [x*x for x in args]
```

```python
result = square(1,2,3,4,5)
result
```

```python
def triple(*args):
  return [x*3 for x in args]
```

```python
result = triple(1,2,3,4,5)
result
```

```python
def square(number):
  return number * number

numbers = [1,2,3,4,5]

mapped_seg = map(square, numbers) # отображает один список на другой по элеменно проход как по циклу
```

```python
for x in mapped_seg:
  print(x)
```

```python
print(type(mapped_seg))
```

```python
list(map(square, numbers))
```

```python
def is_adult(age):
  return age >= 18

ages = [14,18,21,16,30]  
```

```python
filter(is_adult, ages) # Функция filter проверят условие на истинность 
```

```python
list(filter(is_adult, ages))
```

```python
is_adult = lambda age: age >= 18 # Более одной строки быть не может
```

```python
list(filter(is_adult,ages))
```

```python
list(filter(lambda age: age >= 18, ages))
```
