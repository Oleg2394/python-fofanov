```python
def log_decorator(func):
  def wrap():
    print(f"Clling func {func}")
    func()
    print(f"Func {func} finished it's work")
  return wrap   
```

```python
@log_decorator
def hello():
  print("hello, world!")
```

```python
hello()
```

```python
help(hello)
```

```python
from functools import wraps
```

```python
def log_decorator(func):
  @wraps(func)
  def wrap(*args, **kwargs):
    print(f"Clling func {func}")
    func(*args, **kwargs)
    print(f"Func {func} finished it's work")
  return wrap   
```

```python
@log_decorator
def hello():
  print("hello, world!")
```

```python
help(hello)
```
