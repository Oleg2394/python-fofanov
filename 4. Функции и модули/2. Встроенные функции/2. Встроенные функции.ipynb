```python
players = [("Carlsen", 2842), ("Caruana", 2822),
           ("Mamedyarov", 2801), ("Ding", 2797),
           ("Giri", 2780)]
```

```python
all(rating > 2700 for _ , rating in players) # rating больше 2700
```

```python
all(rating > 2800 for _ , rating in players) # rating больше 2800 работает лучше
```

```python
all([rating > 2800 for _ , rating in players]) # rating больше 2800  в виде списка
```
