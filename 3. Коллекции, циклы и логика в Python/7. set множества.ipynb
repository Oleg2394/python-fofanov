```python
set1 = {1,2,3,4}
set2 = {1,2,3,4,5}
```

```python
# Является ли одно множества другим множеством (Включает все элементы первого во второй)
set1.issubset(set2)
```

```python
set2.issubset(set1)
```

```python
set2.issuperset(set1) # Является ли одно множества больше чем другой множеством (Включает все элементы первого во второй)
```

```python
set1 = {1,2,3}
set2 = {4,5,6}
set1.isdisjoint(set2) # Если нет ни одного совпадения то вернут истину
```

```python
set1 = {1,2,3,4}
set2 = {1,2,3,4,5}
set3 = set1.union(set2) # Объединеня двух множеств
set3
```

```python
set1
```

```python
set3 = set1.intersection(set2) # Только одинакове элементы пересечение
set3
```

```python
set1 = {0,1,2,3,4}
set2 = {1,2,3,4,5}
set3 = set1.difference(set2)
set4 = set1.symmetric_difference(set2)
print(set3)
print(set4)
```

```python
set1.update(set2) # Добавить отстоющие элмененты
set1
```

```python
set1.remove(1) # Удалить 1
print(set1)
```

```python
set1.discard(2) # Удалить 2
print(set1)
```

```python
popped_out_element = set1.pop()
print(popped_out_element)
```

```python
set1.clear() # Очистка множества
set1
```
