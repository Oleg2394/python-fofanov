```
names1  = ["John", "Bob", "Alice"]
names2 = ["Tracy", "Elijah", "Mason"]

names_combined = names1 + names2 # Объединение списков 
print(names_combined)
print(type(names_combined))
```

```
names1[0] = "Liam" # Заменяем нулевой индекс
print(names1)
```

```
names1.append("William") # Добавить в конец списка
```

```
names1
```

```
popped = names1.pop() # Удалить последний элемент и вернуть его в перемнную
print(popped)
print(names1)
```

```
names1.pop(0) # Удалить по индексу
print(names1)
```

```
names1.append("James")
names1.sort() # Сортировка по алфавиту
names1
```

```
letters = ['ac', "ab", "aa"]
letters.sort()
print(letters)
```

```
letters = ["abc", "a", "ab"]
letters.sort(key = len) # Сортируем по длине элементов  в строке
print(letters)
```

```
numbers = [3,2,8,5,0,3,4,1,1]
print(numbers)
numbers.sort()
print(numbers)
```

```
numbers = [3,2,8,5,0,3,4,1,1]
numbers.reverse() # Меняем порядок
numbers
```

```
numbers.sort(reverse = True) # Сортировка по убыванию
print(numbers)
```

```
numbers.insert(1,22) # Поставить на место первого элемента 22
print(numbers)
```

```
numbers.index(5) # Найти индекс элемента
```

```
print(numbers.count(3)) # Колисчетво цифр 3 в списке
```

```
copy =  numbers,copy() # Скопировать список
print(copy)
```
