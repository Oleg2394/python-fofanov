```python
import random
```

```python
number = random.randint(1, 50)
```

```python
i = 1
while i <= 6:
  print("Попытка", i)
  number_input = int(input("Введите число в диапазоне от 1 до 50 "))

  if i == 6 and number_input != number:
    print(f"Конец игры загаданное число {number}")
    break

  if number == number_input:
    print(f'Поздравляю победитель ты сделал это за {i} попытки')
    break
  elif number_input < number and 1 <= number_input <= 50:
    print("Загаданное число больше")
    i+=1
    continue
  elif number_input > number and 1 <= number_input <= 50:
    print("Загаданное число меньше")
    i+=1
    continue
  else:
    print("Вы ввели другое число")
print('Конец игры')      
```

```python
import random
tries = 0

number = random.randint(1, 50)
print("Не хотите сыграть в игру? \n угадай число от 1 до 50")

while tries < 6:
  guess = int(input("Введите число "))

  tries += 1

  if guess < number:
    print("Слишком маленькое число")

  if guess > number:
    print("Слишком большое число")   

  if guess == number:
    print(f"Поздравляю победитель, ты использовал {tries} число попыток")
    break
  
  if guess != number and tries == 6:
    print(f"Извините кончились попытки, мое число было ")
    break
```
