```python
# Цикл двойной
list1 = [2,4,-5,6,8,-2]
list2 = [2,-6,8,3,5,-2]

pairs = []
for x in list1:
  for y in list2:
    cur_sum = x + y
    if cur_sum == 0:
      pairs.append((x,y))
pairs      
```
